```python
# def classify_fuel(sample):
def classify_fuel(sample):
# sample = str(sample).lower()
    sample = str(sample).lower()
    if "gumm" in sample or "gm" in sample:
        return "Rubber"
    elif "wod" in sample or 'wood' in sample or 'wd' in sample:
        return "Wood"
    elif "coal" in sample or 'bks' in sample:
        return "Lignite"
    elif "swage" in sample or 'swg' in sample or 'sewage' in sample:
        return "Sewage"
    elif "refuse" in sample or 'ref' in sample:
        return "RDF2"
    elif "digestate" in sample or 'dgst' in sample or "dgt" in sample:
        if all(x not in sample for x in ['pe', 'pp', 'car']):
            return "Digestate"
        elif "pe" in sample:
            return "Digestate_PE"
        elif "pp" in sample:
            return "Digestate_PP"
        elif "car" in sample:
            return "Digestate_SCP"
    elif "htc" in sample:
        return "HTC-MSW"
    elif "cel" in sample:
        return "Cellulose"
    elif "hc" in sample:
        return "Hemicellulose"
    elif "lig" in sample:
        return "Lignin"
    elif "ebs2" in sample:
        return "RDF2"
    elif "ebs1" in sample:
        return "RDF1"
    elif "tc_" in sample:
        return "Torr-Coal"
    elif "tw_" in sample:
        return "Torr-Wood"
    else:
        return "unknown"

# features['fuel_type'] = features['sample'].apply(classify_fuel)
features['fuel_type'] = features['sample'].apply(classify_fuel)

# print(pd.unique(features['fuel_type']))
print(pd.unique(features['fuel_type']))
```

```python
# from scipy import stats
from scipy import stats
# def remove_outliers_z_score(df, threshold=2):
def remove_outliers_z_score(df, threshold=2):
# cleaned = []
    cleaned = []
    for _, grp in df.groupby(['temperature','fuel_type']):
        # if no variation, keep all
        if grp['devol_yield'].nunique() < 2:
            cleaned.append(grp)
            continue
# z = stats.zscore(grp['devol_yield'], nan_policy='omit')
        z = stats.zscore(grp['devol_yield'], nan_policy='omit')
# mask = np.abs(z) > threshold
        mask = np.abs(z) > threshold
        cleaned.append(grp.loc[~mask])
# return pd.concat(cleaned, ignore_index=True)
    return pd.concat(cleaned, ignore_index=True)

# def remove_outliers_modified_z_score(df, threshold=1):
def remove_outliers_modified_z_score(df, threshold=1):
# cleaned = []
    cleaned = []
    for _, grp in df.groupby(['temperature','fuel_type']):
# median = grp['devol_yield'].median()
        median = grp['devol_yield'].median()
# mad    = np.median(np.abs(grp['devol_yield'] - median))
        mad    = np.median(np.abs(grp['devol_yield'] - median))
# if mad == 0:
        if mad == 0:
# mask = np.zeros(len(grp), dtype=bool)
            mask = np.zeros(len(grp), dtype=bool)
        else:
# modz = 0.6745 * (grp['devol_yield'] - median) / mad
            modz = 0.6745 * (grp['devol_yield'] - median) / mad
# mask = np.abs(modz) > threshold
            mask = np.abs(modz) > threshold
        cleaned.append(grp.loc[~mask])
# return pd.concat(cleaned, ignore_index=True)
    return pd.concat(cleaned, ignore_index=True)

# def remove_outliers_mad(df, threshold=2):
def remove_outliers_mad(df, threshold=2):
# cleaned = []
    cleaned = []
    for _, grp in df.groupby(['temperature','fuel_type']):
# median = grp['devol_yield'].median()
        median = grp['devol_yield'].median()
# mad    = np.median(np.abs(grp['devol_yield'] - median))
        mad    = np.median(np.abs(grp['devol_yield'] - median))
# mask   = np.abs(grp['devol_yield'] - median) > threshold * mad
        mask   = np.abs(grp['devol_yield'] - median) > threshold * mad
        cleaned.append(grp.loc[~mask])
# return pd.concat(cleaned, ignore_index=True)
    return pd.concat(cleaned, ignore_index=True)

# def remove_outliers_iqr(df, threshold=1.5):
def remove_outliers_iqr(df, threshold=1.5):
# cleaned = []
    cleaned = []
    for _, grp in df.groupby(['temperature','fuel_type']):
# Q1, Q3 = grp['devol_yield'].quantile([0.25,0.75])
        Q1, Q3 = grp['devol_yield'].quantile([0.25,0.75])
# iqr     = Q3 - Q1
        iqr     = Q3 - Q1
# if iqr == 0:
        if iqr == 0:
            cleaned.append(grp)
        else:
# mask = (grp['devol_yield'] < Q1 - threshold*iqr) | \
            mask = (grp['devol_yield'] < Q1 - threshold*iqr) | \
                   (grp['devol_yield'] > Q3 + threshold*iqr)
            cleaned.append(grp.loc[~mask])
# return pd.concat(cleaned, ignore_index=True)
    return pd.concat(cleaned, ignore_index=True)

# def compare_outlier_methods(df, methods):
def compare_outlier_methods(df, methods):
# stats_list = []
    stats_list = []
    for temp in sorted(df['temperature'].unique()):
# orig_y = df.loc[df['temperature']==temp, 'devol_yield']
        orig_y = df.loc[df['temperature']==temp, 'devol_yield']
# row = {
        row = {
            'temperature':   temp,
            'originalCount': len(orig_y),
            'originalMean':  orig_y.mean(),
            'originalMedian':orig_y.median(),
            'originalStd':   orig_y.std(),
        }
        for name, subdf in methods.items():
# kept = subdf.loc[subdf['temperature']==temp, 'devol_yield']
            kept = subdf.loc[subdf['temperature']==temp, 'devol_yield']
# row[f'{name}KeptCount']  = len(kept)
            row[f'{name}KeptCount']  = len(kept)
# row[f'{name}KeptMean']   = kept.mean()   if len(kept)>0 else np.nan
            row[f'{name}KeptMean']   = kept.mean()   if len(kept)>0 else np.nan
# row[f'{name}KeptMedian'] = kept.median() if len(kept)>0 else np.nan
            row[f'{name}KeptMedian'] = kept.median() if len(kept)>0 else np.nan
# row[f'{name}KeptStd']    = kept.std()    if len(kept)>0 else np.nan
            row[f'{name}KeptStd']    = kept.std()    if len(kept)>0 else np.nan
        stats_list.append(row)
    return stats_list

#Wilcoxon tests
# def test_significance(stats_list):
def test_significance(stats_list):
# from scipy import stats
    from scipy import stats
# df = pd.DataFrame(stats_list)
    df = pd.DataFrame(stats_list)
# methods = ['z', 'modz', 'mad', 'iqr']
    methods = ['z', 'modz', 'mad', 'iqr']
# results = {}
    results = {}
    for i, a in enumerate(methods):
        for b in methods[i+1:]:
# key = f'{a}_vs_{b}'
            key = f'{a}_vs_{b}'
# W, p = stats.wilcoxon(df[f'{a}KeptMean'], df[f'{b}KeptMean'], zero_method='wilcox')
            W, p = stats.wilcoxon(df[f'{a}KeptMean'], df[f'{b}KeptMean'], zero_method='wilcox')
# results[key] = {'W_statistic': W, 'p_value': p}
            results[key] = {'W_statistic': W, 'p_value': p}
    return results


# def calculate_agreement(df, methods):
def calculate_agreement(df, methods):
# flags = {}
    flags = {}
    for name, subdf in methods.items():
# kept_idx = set(subdf.index)
        kept_idx = set(subdf.index)
# flags[name] = df.index.to_series().apply(lambda i: i not in kept_idx)
        flags[name] = df.index.to_series().apply(lambda i: i not in kept_idx)
# agree = {}
    agree = {}
# pairs = [('z','modz'), ('z','mad'), ('z','iqr')]
    pairs = [('z','modz'), ('z','mad'), ('z','iqr')]
    for a, b in pairs:
# agree[f'{a}_vs_{b}'] = (flags[a] == flags[b]).sum()
        agree[f'{a}_vs_{b}'] = (flags[a] == flags[b]).sum()
    return agree



# def main():
def main():
# df = total.copy()
    df = total.copy()
# methods = {
    methods = {
# 'z':    remove_outliers_z_score(df, threshold=1),
        'z':    remove_outliers_z_score(df, threshold=1),
# 'modz': remove_outliers_modified_z_score(df, threshold=1),
        'modz': remove_outliers_modified_z_score(df, threshold=1),
# 'mad':  remove_outliers_mad(df, threshold=2),
        'mad':  remove_outliers_mad(df, threshold=2),
# 'iqr':  remove_outliers_iqr(df, threshold=1.5)}
        'iqr':  remove_outliers_iqr(df, threshold=1.5)}

# comparison_stats = compare_outlier_methods(df, methods)
    comparison_stats = compare_outlier_methods(df, methods)
# comparison_df    = pd.DataFrame(comparison_stats)
    comparison_df    = pd.DataFrame(comparison_stats)
# significance = test_significance(comparison_stats)
    significance = test_significance(comparison_stats)
# agreement = calculate_agreement(df, methods)
    agreement = calculate_agreement(df, methods)
# print("--- Comparison Stats ---")
    print("--- Comparison Stats ---")
# print(comparison_df)
    print(comparison_df)

# print("\n--- Significance Tests ---")
    print("\n--- Significance Tests ---")
# print(significance)
    print(significance)

# print("\n--- Agreement Metrics ---")
    print("\n--- Agreement Metrics ---")
# print(agreement)
    print(agreement)

# if __name__ == "__main__":
if __name__ == "__main__":
    main()
```

```python
# def remove_outliers_modified_z_score(df, threshold=1):
def remove_outliers_modified_z_score(df, threshold=1):
# result = []
    result = []
    for temp, group in df.groupby(['temperature', 'fuel_type', 'residence_time']):
# median = group['devol_yield'].median()
        median = group['devol_yield'].median()
# mad = np.median(np.abs(group['devol_yield'] - median))
        mad = np.median(np.abs(group['devol_yield'] - median))
# if mad == 0:
        if mad == 0:
# modified_z = np.zeros_like(group['devol_yield'])
            modified_z = np.zeros_like(group['devol_yield'])
        else:
# modified_z = 0.6745 * (group['devol_yield'] - median) / mad
            modified_z = 0.6745 * (group['devol_yield'] - median) / mad
# group = group.copy()
        group = group.copy()
# group['isOutlierModZ'] = np.abs(modified_z) > threshold
        group['isOutlierModZ'] = np.abs(modified_z) > threshold
# group = group[~group['isOutlierModZ']]  # remove outliers
        group = group[~group['isOutlierModZ']]  # remove outliers
        result.append(group)
# return pd.concat(result).drop(columns='isOutlierModZ')
    return pd.concat(result).drop(columns='isOutlierModZ')


# def remove_outliers_mad(df, threshold=1):
def remove_outliers_mad(df, threshold=1):
# result = []
    result = []
    for temp, group in df.groupby(['temperature', 'fuel_type', 'residence_time']):
# median = group['devol_yield'].median()
        median = group['devol_yield'].median()
# mad = np.median(np.abs(group['devol_yield'] - median))
        mad = np.median(np.abs(group['devol_yield'] - median))
# outliers = np.abs(group['devol_yield'] - median) > threshold * mad
        outliers = np.abs(group['devol_yield'] - median) > threshold * mad
# group = group.copy()
        group = group.copy()
# group['isOutlierMAD'] = outliers
        group['isOutlierMAD'] = outliers
# group = group[~group['isOutlierMAD']]  # remove outliers
        group = group[~group['isOutlierMAD']]  # remove outliers
        result.append(group)
# return pd.concat(result).drop(columns='isOutlierMAD')
    return pd.concat(result).drop(columns='isOutlierMAD')

# --- IQR Outlier Removal ---
# def remove_outliers_iqr(df, threshold=1.5):
def remove_outliers_iqr(df, threshold=1.5):
# cleaned = []
    cleaned = []
    for _, grp in df.groupby(['temperature', 'fuel_type']):
# Q1, Q3 = grp['devol_yield'].quantile([0.25, 0.75])
        Q1, Q3 = grp['devol_yield'].quantile([0.25, 0.75])
# IQR      = Q3 - Q1
        IQR      = Q3 - Q1
# mask_out = (grp['devol_yield'] < Q1 - threshold*IQR) | \
        mask_out = (grp['devol_yield'] < Q1 - threshold*IQR) | \
                   (grp['devol_yield'] > Q3 + threshold*IQR)
# grp = grp.loc[~mask_out]            # ❶ actually drop them
        grp = grp.loc[~mask_out]            # ❶ actually drop them
        cleaned.append(grp)
# return pd.concat(cleaned, ignore_index=True)
    return pd.concat(cleaned, ignore_index=True)

# --- Z-score Outlier Removal ---
# def remove_outliers_z_score(df, threshold=3):
def remove_outliers_z_score(df, threshold=3):
# cleaned = []
    cleaned = []
    for _, grp in df.groupby(['temperature', 'fuel_type']):
# z = stats.zscore(grp['devol_yield'], nan_policy="omit")
        z = stats.zscore(grp['devol_yield'], nan_policy="omit")
# mask_out = np.abs(z) > threshold
        mask_out = np.abs(z) > threshold
# grp = grp.loc[~mask_out]            # ❶ actually drop them
        grp = grp.loc[~mask_out]            # ❶ actually drop them
        cleaned.append(grp)
# return pd.concat(cleaned, ignore_index=True)
    return pd.concat(cleaned, ignore_index=True)

# def remove_outliers(df, method="mad", **kwargs):
def remove_outliers(df, method="mad", **kwargs):
# funcs = {
    funcs = {
        "mad":   remove_outliers_mad,
        "modz":  remove_outliers_modified_z_score,
        "iqr":   remove_outliers_iqr,
        "z":     remove_outliers_z_score,
    }
    return funcs[method](df, **kwargs)
```
